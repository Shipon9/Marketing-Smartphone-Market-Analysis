# Parsing
Data were gathered from Amazon.com.

The are 2 files:

1) File with phone models name and ID

2) File with customers reviews of phones


# Import libraries and parsed data 

```python
import json
import requests
import urllib
import pandas as pd
```

```python
# Read first file
targetUrl = "https://disk.yandex.ru/i/kLIWk-Hl4xywLQ" 
url = "https://cloud-api.yandex.net/v1/disk/public/resources?public_key=" + urllib.parse.quote(targetUrl, safe="") 
responseData = requests.get(url) 
url = responseData.json()['file'] + '/' + responseData.json()['name'] 
models_name = pd.read_excel(url)
```

```python
models_name
```

```python
#Read second file
targetUrl = "https://disk.yandex.ru/i/evtSY1UlgIwYkA" 
url = "https://cloud-api.yandex.net/v1/disk/public/resources?public_key=" + urllib.parse.quote(targetUrl, safe="") 
responseData = requests.get(url) 
url = responseData.json()['file'] + '/' + responseData.json()['name'] 
all_models = pd.read_excel(url)
```

```python
all_models
```

```python
#checking the data type of the columns

all_models.info()
```

```python
#Converting columns "Content" and "Title" to string dtype

all_models['Content'] = all_models['Content'].astype(str).str.lower()
all_models['Title'] = all_models['Title'].astype(str).str.lower()
```

```python
#Tokenization

from nltk.tokenize import RegexpTokenizer

regexp = RegexpTokenizer('\w+')

all_models['Content_token']=all_models['Content'].apply(regexp.tokenize)
all_models.head(3)
```

```python
#In case we want to do it for the column "Title as well"

#all_models['Title_token'] = all_models['Title'].apply(regexp.tokenize)

#all_models.head(3)
```

```python
#Stopwords

import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords

# Make a list of english stopwords
stopwords = nltk.corpus.stopwords.words("english")
```

```python
#We use a lambda function to remove the stopwords:
# Remove stopwords

all_models['Content_token'] = all_models['Content_token'].apply(lambda x: [item for item in x if item not in stopwords])
all_models.head(3)
```

```python
#In case we want to do the same for "Title" column

#all_models['Title_token'] = all_models['Title_token'].apply(lambda x: [item for item in x if item not in stopwords])
#all_models.head(3)
```

```python
#Removing infrequent words
#We first change the format of text_token to strings and keep only words which are longer than 2 letters

all_models['Content_string'] = all_models['Content_token'].apply(lambda x: ' '.join([item for item in x if len(item)>2]))
```

```python
all_models[['Content', 'Content_token', 'Content_string']].head()
```

```python
#Creating a list of all words

all_words = ' '.join([word for word in all_models['Content_string']])
```

```python
#Tokenize all_words

import nltk
nltk.download('punkt')

tokenized_words = nltk.tokenize.word_tokenize(all_words)
```

```python
#Create a frequency distribution which records the number of times each word has occurred:

from nltk.probability import FreqDist

fdist = FreqDist(tokenized_words)
fdist
```

```python
#we can use our fdist dictionary to drop words which occur less than a certain amount of times

all_models['Content_string_fdist'] = all_models['Content_token'].apply(lambda x: ' '.join([item for item in x if fdist[item] >= 4 ]))
```

```python
all_models[['Content', 'Content_token', 'Content_string', 'Content_string_fdist']].head()
```

```python
#Lemmatization

#Next, we perfom lemmatization.

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer

wordnet_lem = WordNetLemmatizer()

all_models['Content_string_lem'] = all_models['Content_string_fdist'].apply(wordnet_lem.lemmatize)
```

```python
#Sentiment analysis

#VADER lexicon

nltk.download('vader_lexicon')
```

```python
#Sentiment

#Sentiment Intensity Analyzer

from nltk.sentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

#Polarity scores
#Use the polarity_scores method:

all_models['polarity'] = all_models['Content_string_lem'].apply(lambda x: analyzer.polarity_scores(x))
all_models.tail(3)
```

```python
#Transform data
# Change data structure
all_models = pd.concat(
    [all_models.drop(['Title', 'Title_URL', 'Image', 'aprofile_URL', 'View', 'acolorsecondary', 'State', 'asizebase', 'asizebase_URL', \
                      'asizebase1', 'abuttontext_URL', 'polarity'], axis=1), 
     all_models['polarity'].apply(pd.Series)], axis=1)

all_models.head(3)
```

```python
# Create new variable with sentiment "neutral," "positive" and "negative"
all_models['sentiment'] = all_models['compound'].apply(lambda x: 'positive' if x >0 else 'neutral' if x==0 else 'negative')
all_models.head(4)
```

```python
all_models['rating'] = all_models['aiconalt'].apply(lambda x: int(x[:1]))
all_models.head(4)
```
